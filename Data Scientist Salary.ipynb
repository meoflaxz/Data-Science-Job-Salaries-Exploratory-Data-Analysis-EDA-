```python
job_v_salary = data.drop(columns=["experience_level", "employment_type", "employee_residence", "company_location"], axis=1)
job_v_salary = job_v_salary.sort_values('salary_in_usd', ascending=False)
#job_v_salary = job_v_salary.head(30)
job_v_salary
```

```python
grouped = job_v_salary.loc[:,['job_title', 'salary_in_usd']].groupby(['job_title']).median().sort_values(by='salary_in_usd', ascending=False)
```

```python
#sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(40,20))
sns.boxplot(data = job_v_salary, x=job_v_salary.salary_in_usd, y=job_v_salary.job_title, orient='h', ax=ax, order=grouped.index)
plt.title("Boxplot of Data Scientist Salary", fontsize=20)
plt.xlabel("Salary (USD)", fontsize=20)
plt.ylabel("Job Title", fontsize=20)
plt.show()
```

```python
# Average salary in usd for each job title
avg_job_salary = pd.DataFrame(data.groupby("job_title")['salary_in_usd'].mean('salary_in_usd').sort_values(ascending=False).head(20))
avg_job_salary = avg_job_salary.reset_index()
avg_job_salary['salary_in_usd'] = round(avg_job_salary['salary_in_usd'], 0)
#avg_job_salary['salary_in_usd'] = avg_job_salary["salary_in_usd"].map('{:,.0f}'.format)
avg_job_salary
```

```python
job_title = avg_job_salary['job_title']
salary_in_usd = avg_job_salary['salary_in_usd']
```

```python
plt.figure(figsize=(20,10))

bar = sns.barplot(salary_in_usd, job_title,palette='flare')
plt.title("Top 20 Average Salary of Data Scientist by Job Title", fontsize= 20)
plt.ylabel("Job Title", fontsize= 15)
plt.xlabel("Average Salary (USD)", fontsize= 15)
plt.bar_label(bar.containers[0],size=12)
plt.show()
```

```python
avg_salary_company = pd.DataFrame(data.groupby("company_location")["salary_in_usd"].mean().sort_values(ascending=False).head(20))
avg_salary_company = avg_salary_company.reset_index()
avg_salary_company['salary_in_usd'] = round(avg_salary_company['salary_in_usd'], 0)
avg_salary_company
```

```python
company_location = avg_salary_company['company_location']
salary_in_usd = avg_salary_company['salary_in_usd']
```

```python
country = ['Russia', 'USA', 'New Zealand', "Israel", 'Japan', 'Australia', 'Iraq', 'United Arab Emirates','Canada', 'Singapore', 'Belgium', 'Germany', 'United Kingdom', 'Austria', \
            'China', 'Ireland', 'Czechia', 'Denmark', 'Poland', 'Switzerland']

plt.figure(figsize=(20,10))
bar = sns.barplot( salary_in_usd, company_location,palette='crest')
plt.title("Average Salary of Data Scientist by Country", fontsize= 20)
plt.ylabel("Country", fontsize= 15)
plt.xlabel("Average Salary (USD)", fontsize= 15)
plt.bar_label(bar.containers[0],size=12)
bar.set_yticklabels(country)
plt.show()
```

```python
job_title_list = data.loc[:,'job_title']
```

```python
top_job = pd.DataFrame(job_title_list.value_counts().head(20))
top_ten_job = pd.DataFrame(job_title_list.value_counts().head(8))
```

```python
top_job = top_job.reset_index()
top_ten_job = top_ten_job.reset_index()
```

```python
top_job.dtypes
```

```python
top_job.info()
```

```python
job = top_job['index']
frequency = top_job['job_title']

ten_job = top_ten_job['index']
ten_frequency = top_ten_job['job_title']
```

```python
plt.figure(figsize=(20,10))
bar = sns.barplot(frequency, job, palette='viridis')
plt.bar_label(bar.containers[0], size=15)
plt.title("Top 20 Most Popular Job Title", fontsize= 20)
plt.xlabel("Frequency of Job", fontsize = 15)
plt.ylabel("Job Title", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(10,10))
colors = sns.color_palette('pastel')

plt.pie(ten_frequency, labels=ten_job, colors=colors, autopct='%.0f%%')
plt.title('Top 8 Most Popular Job Title In Data Science Domain', fontsize=20)
plt.show()
```
